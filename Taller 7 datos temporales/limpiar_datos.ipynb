```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('ML-Dataset.csv')

# Visualizar los tipos de datos originales
print("Tipos de datos antes de la limpieza:")
print(df.dtypes[['OrderDate', 'EmployeeHireDate', 'RegionName', 'PerUnitPrice']])
```

```python
# 1. Convertir fechas a objetos Datetime
# El formato en tu CSV es día-mes corto-año (ej. 17-Nov-16)
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%d-%b-%y')
df['EmployeeHireDate'] = pd.to_datetime(df['EmployeeHireDate'], format='%d-%b-%y')

# 2. Limpieza de strings (Trim y eliminación de saltos de línea)
# Aplicamos a todas las columnas de tipo objeto (texto)
text_columns = df.select_dtypes(include=['object']).columns
for col in text_columns:
    df[col] = df[col].astype(str).str.strip() # Quita espacios al inicio/final
    df[col] = df[col].str.replace(r'\r+|\n+', ' ', regex=True) # Quita saltos de línea

# 3. Asegurar valores numéricos
# Forzamos a que sean numéricos, si hay errores los convierte en NaN
cols_numericas = ['ProductStandardCost', 'Profit', 'ProductListPrice', 'PerUnitPrice']
for col in cols_numericas:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("¡Limpieza completada!")
```

```python
# Tarea 2: Extraer el Trimestre (Simulando DATEPART)
df['OrderQuarter'] = df['OrderDate'].dt.quarter

# Tarea 1: Extraer Día de la Semana (Para el Heatmap cíclico)
# 0=Lunes, 6=Domingo
df['OrderDayName'] = df['OrderDate'].dt.day_name()

# Tarea 3: Truncar fecha a mes (Para el crecimiento de empleados)
df['HireMonthTrunc'] = df['EmployeeHireDate'].dt.to_period('M').dt.to_timestamp()

# Guardar el archivo limpio para el Dashboard
df.to_csv('ML-Dataset_Clean.csv', index=False)
print("Archivo 'ML-Dataset_Clean.csv' guardado y listo para el dashboard.")
```
